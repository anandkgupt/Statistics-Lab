```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the parameters of the discrete uniform distribution
low = 1  # Lower bound (inclusive)
high = 6  # Upper bound (inclusive)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Generate values for the x-axis (possible outcomes)
x = np.arange(low, high + 1)

# Calculate the PMF (probability mass function)
pmf = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title(f'PMF of Discrete Uniform Distribution ({low}, {high})')
plt.grid(True)
plt.show()
```

Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import numpy as np

def bernoulli_pdf(x, p):
  """
  Calculates the probability density function (PDF) of a Bernoulli distribution.

  Args:
    x: The value of the random variable (0 or 1).
    p: The probability of success (probability of x=1).

  Returns:
    The probability density function value.
  """

  if x not in [0, 1]:
    raise ValueError("x must be either 0 or 1")

  if not (0 <= p <= 1):
    raise ValueError("p must be between 0 and 1")

  if x == 1:
    return p
  else:
    return 1 - p

# Example usage:
p = 0.7  # Probability of success
x = 1     # Value of the random variable

pdf_value = bernoulli_pdf(x, p)
print(f"PDF of Bernoulli({p}) at x={x}: {pdf_value}")
```

Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
  """
  Calculates Z-scores for a given dataset.

  Args:
    data: A NumPy array or list containing the data.

  Returns:
    A NumPy array of Z-scores.
  """
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

# Example usage
data = np.array([10, 12, 8, 15, 11, 9, 13])
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(8, 4))
plt.bar(range(len(data)), z_scores, color='skyblue')
plt.xlabel('Data Points')
plt.ylabel('Z-Score')
plt.title('Z-Scores of the Dataset')
plt.grid(True)
plt.show()
```

Q18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_score(data_point, mean, std_dev):
  """
  Calculates the Z-score for a given data point and plots it on a standard normal distribution.

  Args:
    data_point: The data point for which to calculate the Z-score.
    mean: The mean of the population.
    std_dev: The standard deviation of the population.
  """

  # Calculate the Z-score
  z_score = (data_point - mean) / std_dev

  # Generate x-values for plotting the standard normal distribution
  x = np.linspace(-3, 3, 100)
  pdf = norm.pdf(x, 0, 1)

  # Plot the standard normal distribution
  plt.figure(figsize=(8, 5))
  plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')

  # Plot a vertical line at the Z-score
  plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')

  # Add labels and title
  plt.xlabel('Z-score')
  plt.ylabel('Probability Density')
  plt.title('Standard Normal Distribution with Z-score')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example usage
data_point = 75
mean = 70
std_dev = 5

calculate_and_plot_z_score(data_point, mean, std_dev)
```

Q19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
  """
  Performs a one-sample Z-test.

  Args:
    sample_mean: The mean of the sample.
    population_mean: The hypothesized population mean.
    population_std_dev: The population standard deviation.
    sample_size: The size of the sample.
    alpha: Significance level (default: 0.05).

  Returns:
    A tuple containing:
      - z_score: The calculated Z-score.
      - p_value: The p-value of the test.
      - reject_null: True if the null hypothesis is rejected, False otherwise.
  """

  # Calculate the Z-score
  z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

  # Calculate the p-value
  p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

  # Determine if the null hypothesis is rejected
  reject_null = p_value < alpha

  return z_score, p_value, reject_null

# Example usage
sample_mean = 102  # Sample mean
population_mean = 100  # Hypothesized population mean
population_std_dev = 15  # Population standard deviation
sample_size = 50

z_score, p_value, reject_null = z_test(sample_mean, population_mean, population_std_dev, sample_size)

print(f"Z-score: {z_score:.2f}")
print(f"p-value: {p_value:.4f}")

if reject_null:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

Q20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

# Generate a sample dataset (e.g., random data for demonstration)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std Dev=10, Sample Size=100

# Calculate the sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
sample_size = len(data)
standard_error = sample_std / np.sqrt(sample_size)

# Confidence level (e.g., 95%)
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for 95% confidence

# Calculate the margin of error and confidence interval
margin_of_error = z_critical * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Z-critical value: {z_critical:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
```

Q21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

# Generate random data from a normal distribution
mu = 50  # Population mean
sigma = 10  # Population standard deviation
sample_size = 30
data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation

# Confidence level
confidence_level = 0.95

# Calculate degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate t-critical value
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

# Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are {confidence_level * 100}% confident that the true population mean falls within the interval "
      f"({lower_bound:.2f}, {upper_bound:.2f}).")
```

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parameters for the Poisson distribution
lambda_value = 4  # Average rate (mean)

# Define the range of possible values (0 to some maximum)
x_values = np.arange(0, 11)  # Values from 0 to 10

# Calculate the CDF for each value
cdf_values = poisson.cdf(x_values, mu=lambda_value)

# Display the CDF values
for x, cdf in zip(x_values, cdf_values):
    print(f"P(X <= {x}) = {cdf:.4f}")

# Plot the CDF
plt.step(x_values, cdf_values, where='post', label="CDF", color='blue')
plt.title("CDF of Poisson Distribution (Œª=4)")
plt.xlabel("x (Number of Events)")
plt.ylabel("P(X ‚â§ x)")
plt.xticks(x_values)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

Q25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

def compare_standard_deviations(data1, data2):
  """
  Compares the standard deviations of two datasets.

  Args:
    data1: The first dataset.
    data2: The second dataset.

  Returns:
    A tuple containing:
      - std_dev1: Standard deviation of the first dataset.
      - std_dev2: Standard deviation of the second dataset.
  """

  std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation
  std_dev2 = np.std(data2, ddof=1)  # Sample standard deviation

  return std_dev1, std_dev2

# Sample datasets
data1 = np.random.normal(loc=10, scale=2, size=100)  # Mean 10, Standard Deviation 2
data2 = np.random.normal(loc=15, scale=5, size=100)  # Mean 15, Standard Deviation 5

# Calculate standard deviations
std_dev1, std_dev2 = compare_standard_deviations(data1, data2)

# Print results
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Visualize the data (optional)
plt.figure(figsize=(8, 4))
plt.hist(data1, bins=20, alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.title('Comparison of Datasets')
plt.show()

# Perform F-test to compare variances
f_statistic, p_value = f_oneway(data1, data2)

# Significance level
alpha = 0.05

if p_value < alpha:
    print(f"F-test: Reject the null hypothesis (variances are significantly different).")
else:
    print(f"F-test: Fail to reject the null hypothesis (variances may be equal).")
```

Q26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

# Generate sample data from a normal distribution
mu = 0  # Mean of the distribution
sigma = 1  # Standard deviation of the distribution
sample_size = 100
data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Calculate range
data_sorted = np.sort(data)
data_range = data_sorted[-1] - data_sorted[0]

# Calculate IQR
q1, q3 = np.percentile(data, [25, 75])  # Calculate Q1 and Q3 simultaneously
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

Q27. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (standard normal distribution)
data = np.random.randn(1000)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Perform Z-score normalization
z_scores = (data - mean) / std_dev

# Create histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.7, label='Original Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Original Data')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, density=True, alpha=0.7, label='Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Histogram of Z-scores')

plt.legend()
plt.tight_layout()
plt.show()
```

Q28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np

def calculate_skewness_kurtosis(data):
  """
  Calculates the skewness and kurtosis of a given dataset.

  Args:
    data: A NumPy array containing the dataset.

  Returns:
    A tuple containing:
      - skewness: The skewness of the dataset.
      - kurtosis: The kurtosis of the dataset (Fisher's definition).
  """

  skewness = np.mean(((data - np.mean(data)) / np.std(data))**3)
  kurtosis = np.mean(((data - np.mean(data)) / np.std(data))**4) - 3

  return skewness, kurtosis

# Generate sample data from a normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
sample_size = 1000
data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")
```
